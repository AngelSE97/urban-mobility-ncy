```python
import pandas as pd

path = "/content/drive/MyDrive/Colab_Notebooks/urban_mobility_project/transformed/weather_transformed.parquet"
weather_transformed = pd.read_parquet(path)
path = "/content/drive/MyDrive/Colab_Notebooks/urban_mobility_project/transformed/accidents_transformed.parquet"
accidents_transformed = pd.read_parquet(path)
path = "/content/drive/MyDrive/Colab_Notebooks/urban_mobility_project/transformed/traffic_transformed.parquet"
traffic_transformed = pd.read_parquet(path)
```

```python
#Filtrado de datasets y agregacion de index
weather_2025 = weather_transformed.reset_index(drop=True)

weather_2025['id'] = weather_2025.index + 1

if 'id' in weather_2025.columns:
    cols = ['id'] + [c for c in weather_2025.columns if c != 'id']
    weather_2025 = weather_2025[cols]
```

```python
#Filtrado de datasets y agregacion de index
traffic_daily = traffic_transformed.groupby(['date', 'Boro','street','fromSt','toSt','Direction'],as_index=False).agg({'Vol': 'sum'})
traffic_2025 = traffic_daily[traffic_daily['date'].dt.year == 2025]

traffic_2025 = traffic_2025.reset_index(drop=True)

traffic_2025['id'] = traffic_2025.index + 1

if 'id' in traffic_2025.columns:
    cols = ['id'] + [c for c in traffic_2025.columns if c != 'id']
    traffic_2025 = traffic_2025[cols]



```

```python
#Filtrado de datasets y agregacion de index
cols = [
    'CRASH_DATE',
    'UNIQUE_ID',
    'CRASH_TIME',
    'STATE_REGISTRATION',
    'VEHICLE_TYPE',
    'VEHICLE_MAKE',
    'VEHICLE_MODEL',
    'VEHICLE_YEAR',
    'PRE_CRASH',
    'POINT_OF_IMPACT'
]
accidents_2025 = accidents_transformed[
    (accidents_transformed['CRASH_DATE'].dt.year == 2025) &
    (accidents_transformed['STATE_REGISTRATION'] == 'NY')
][cols].reset_index(drop=True)

accidents_2025 = accidents_2025.reset_index(drop=True)

accidents_2025['id'] = accidents_2025.index + 1

if 'id' in accidents_2025.columns:
    cols = ['id'] + [c for c in accidents_2025.columns if c != 'id']
    accidents_2025 = accidents_2025[cols]
```

```python
weather_curated = weather_2025
accidents_curated = accidents_2025
traffic_curated = traffic_2025
```

```python
weather_curated.to_parquet("/content/drive/MyDrive/Colab_Notebooks/urban_mobility_project/curated/weather_curated.parquet", index=False)
accidents_curated.to_parquet("/content/drive/MyDrive/Colab_Notebooks/urban_mobility_project/curated/accidents_curated.parquet", index=False)
traffic_curated.to_parquet("/content/drive/MyDrive/Colab_Notebooks/urban_mobility_project/curated/traffic_curated.parquet", index=False)
```
